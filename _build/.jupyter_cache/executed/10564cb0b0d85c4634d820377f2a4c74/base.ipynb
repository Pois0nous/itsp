```python
import numpy as np
import matplotlib.pyplot as plt

B = np.random.randn(2,2)
x = np.matmul(np.random.randn(100,2),B)

X = x[:,0:1].T
Y = x[:,1:2].T

A = np.matmul(
    np.linalg.inv(
        np.matmul(X,X.T) ),
    np.matmul(X,Y.T) )

Yhat = np.matmul(A,X)

plt.scatter(X,Y,label='True')
plt.scatter(X,Yhat,label='Estimated')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title('Estimating Y from X')
plt.legend()
plt.show()
```
